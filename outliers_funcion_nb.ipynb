```python
def count_outliers(df):
    
    """ Esta función cuenta el número de outliers para cada columna

    Returns:
        _type_: Devuelve un dataframe 
    """
    counts = df.apply(lambda column: (column == 'atipico').sum())
    return pd.DataFrame({'Outliers_Count': counts})
```

```python
outliers_count_df = count_outliers(outliers_st)

outliers_count_df
```

```python
outliers_st = outliers_st.reset_index()
```

Buscamos un procedimiento para agrupar cuales son las ciudades que son outliers en cada caracteristicas, para saber el número, y cuáles son.

```python
# Se crea un diccionario vacío
outliers_cases = {}

# Se itera sobre cada columna
for column in outliers_st.columns:
    # expresion para crear una lista de los elementos ciudades que cumplen la condición de ser un dato atipico
    outliers_cases[column] = list(outliers_st[outliers_st[column] == 'atipico']['Ciudades'])

print(outliers_cases)
```

```python
for variable, ciudades in outliers_cases.items():
    print(f"Atipicos for {variable}: {ciudades}")
```
